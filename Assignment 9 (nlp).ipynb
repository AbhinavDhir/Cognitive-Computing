```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.probability import FreqDist 
import string 
# 1.1
text = """Wondering how to make ice cream at home? We’ve got you covered. Here’s a brief overview of what you can expect when you make vanilla ice cream at home: 
Cook the milk, cream, and sugar until the sugar has dissolved.Transfer the mixture to a measuring cup, then add the vanilla extract. Chill overnight.Pour the mix into an ice cream maker and churn according to the manufacturer's directions.Serve immediately or ripen in the freezer before serving.
"""

a = text.lower() 
print(a)  

import re 
para = re.sub(r'[^\w\s]','', text ) 
print(para)                             





```

```python
#1.2
words =word_tokenize(text) 
print(words) 

sentences = sent_tokenize(text) 
print(sentences) 
```

```python
#1.3
stop_words= set(stopwords.words('english')) 

filtered_sentence = [word for word in words if word.lower() not in stop_words] 
print(" ".join(filtered_sentence)) 
```

```python
#1.4
freq_dist = FreqDist(filtered_sentence) 
print(freq_dist)
import matplotlib.pyplot as plt 
freq_dist.plot() 
```

```python
#2.1 

from nltk.stem import PorterStemmer 
ps = PorterStemmer()
from nltk.stem import LancasterStemmer 
ls = LancasterStemmer()



for w in filtered_sentence:
    print(w,":",ps.stem(w)) 



```

```python
for w in filtered_sentence:
    print(w,":",ls.stem(w)) 
```

```python
from nltk.stem import WordNetLemmatizer 
lem = WordNetLemmatizer()

for w in filtered_sentence:
    print(w,":",lem.lemmatize(w))
```

```python
porter_stems = [ps.stem(word) for word in filtered_sentence] 
lancaster_stems = [ls.stem(word) for word in filtered_sentence]
lemmatized_words = [lem.lemmatize(word) for word in filtered_sentence] 

import pandas as pd 
data ={
    "original words": filtered_sentence,
    "PorterStemmer": porter_stems,
    "LancasterStemmer":lancaster_stems, 
    "lemmatizedStemmer":lemmatized_words
}
df = pd.DataFrame(data) 
print(df) 
```

```python
def custom_tokenizer(text):
    text = re.sub(r"[^\w\s'-]", "", text)
    tokens = re.findall(r"\b\w+(?:-\w+)*\b", text)
    return tokens

original_text = """
Hi there! You can reach out to us via email at support@company.com or sales@company.com. 
For immediate assistance, call our helpline at 123-456-7890 or +44 20 7946 0958. 
Visit our website at https://www.company.com or follow us on social media at https://twitter.com/company. 
Remember to check out our latest offers and discounts! Thank you.
"""

tokens = custom_tokenizer(original_text)
print("Tokens:", tokens) 
print()
tokeno = custom_tokenizer(text)  
print("text: ",tokeno) 
print()

def clean_text_with_placeholders(text):
    text = re.sub(r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b", "<EMAIL>", text)
    text = re.sub(r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+", "<URL>", text)
    text = re.sub(r"\b\d{3}[-.\s]?\d{3}[-.\s]?\d{4}\b", "<PHONE>", text)
    text = re.sub(r"\+?\d{1,3}[-.\s]?\d{1,4}[-.\s]?\d{1,4}[-.\s]?\d{1,9}\b", "<PHONE>", text)
    return text
print(original_text) 
print()
cleaned_text = clean_text_with_placeholders(original_text)
print("Cleaned Text:", cleaned_text)
    

```
