```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
#4.1
ai_text = "Artificial Intelligence (AI) enables machines to learn from data and make decisions. It powers applications like voice recognition, autonomous vehicles, and personalized recommendations."
bc_text= "Blockchain is a distributed ledger technology that ensures secure and transparent transactions. It is the backbone of cryptocurrencies and enables smart contract execution."

tokens_text1 = set(word_tokenize(ai_text.lower()))
tokens_text2 = set(word_tokenize(bc_text.lower()))

#4.2
intersection = tokens_text1.intersection(tokens_text2)
union = tokens_text1.union(tokens_text2)
jaccard_similarity = len(intersection) / len(union)

#4.3
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([ai_text, bc_text])
cosine_sim = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]

#4.4
print(f"Jaccard Similarity: {jaccard_similarity:.4f}")
print(f"Cosine Similarity: {cosine_sim:.4f}")
```

```python
#5.1

from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt

review = "This ice cream is absolutely delightful! The creamy texture and rich flavors make it a treat. I would highly recommend it to anyone with a sweet tooth."

blob = TextBlob(review)
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity
#5.2

if polarity > 0:
    sentiment = "Positive"
elif polarity < 0:
    sentiment = "Negative"
else:
    sentiment = "Neutral"

print(f"Review: {review}")
print(f"Polarity: {polarity:.2f}, Subjectivity: {subjectivity:.2f}")
print(f"Sentiment: {sentiment}")

#5.3

if sentiment == "Positive":
    wordcloud = WordCloud(width=800, height=400, background_color="white").generate(review)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title("Word Cloud for Positive Review")
    plt.show()
```

```python
import numpy as np
from keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense



####-------------tenserflow not supported------------------------#### (code given below) ----- run on older version--------------

text = ("Ice cream is a delightful treat loved by people of all ages. "
        "Its creamy texture, rich flavors, and refreshing chill make it perfect for any occasion. "
        "Whether enjoyed in a cone, a cup, or as part of a sundae, ice cream brings joy and comfort. "
        "It comes in countless flavors, from classic vanilla and chocolate to exotic mango and pistachio.")

#6.1
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1
input_sequences = []

for line in text.split("."):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i + 1]
        input_sequences.append(n_gram_sequence)

# pad_sequnce
max_sequence_len = max(len(x) for x in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')


X, y = input_sequences[:, :-1], input_sequences[:, -1]
y = np.array(y)

# 6.2
model = Sequential([
    Embedding(total_words, 10, input_length=max_sequence_len - 1),
    LSTM(100),
    Dense(total_words, activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, verbose=0)

# 6.3
seed_text = "Ice cream"
next_words = 10

for _ in range(next_words):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences([token_list], maxlen=max_sequence_len - 1, padding='pre')
    predicted = model.predict(token_list, verbose=0).argmax()
    output_word = ""
    for word, index in tokenizer.word_index.items():
        if index == predicted:
            output_word = word
            break
    seed_text += " " + output_word

print("Generated Text:")
print(seed_text)
```
